# Scikit-Learn 기반 Framework

## Estimator - fit(), predict()

지도학습에서는 분류(Classification), 회귀(Regression) 문제를 다루게 됩니다.<br>
Scikit-Learn에서 제공하는 ML 알고리즘은 모두 분류와 회귀를 지원하며, 이 모든 알고리즘들의 클래스들을 통틀어 Estimator라고 부릅니다.

모든 Estimator는 `fit()`, `predict()` 함수를 기본적으로 가지고 있습니다.<br>
교차검증(Cross Validation)과 같은 Evaluation 함수나 GridSearch 같은 하이퍼파라미터 튜닝을 지원하는 클래스의 경우, 이 Estimator를 인자로 받게 됩니다.<br>
Evaluation 함수나 하이퍼파라미터 튜닝 클래스에 있는 `fit()`, `predict()`는 이 Estimator의 동일함수를 실행하게 됩니다.

### 분류 (Classification)의 종류

- DecisionTreeClassifier
- RandomForestClassifier
- GraidentBoostingClassifer
- GaussianNB
- SVC

### 회귀 (Regression)의 종류

- LinearRegression
- Ridge
- Lasso
- RandomForestRegressor
- GradientBoostingRegressor
