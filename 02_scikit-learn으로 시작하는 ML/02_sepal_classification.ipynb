# 붓꽃 품종 예측하기

첫 번째로 만들 모델은 붓꽃 데이터 세트로 품종을 분류하는 모델입니다.<br>
분류(Classification)는 지도학습(Supervised Learning) 방법 중의 하나로서,<br>
다양한 Feature 값과 Label 값으로 모델을 학습한 뒤, 테스트 데이터에서 Label을 예측하는 형태입니다.

Kaggle에서는 보통 데이터의 구성을 다음과 같이 제공합니다.
- train: 학습에 쓰일 데이터 셋 (Label 있음)
- val: 모델 검증용 데이터 셋 (Label 있음)
- test: 예측용 데이터 셋 (Label 없음)

따라서 다른 ML을 모델을 만들 때 해당 데이터 셋 구성으로 준비해두는 것이 좋습니다.<br>
물론 test 데이터 셋은 자율적으로 구성해도 됩니다.

## 데이터 로드 및 모델 선택

붓꽃 데이터는 Scikit-Learn 패키지에서 불러올 수 있습니다.<br>
데이터는 `sklearn.datasets` 안에 있습니다.

ML 알고리즘의 경우 DecisionTreeClassifier를 사용하겠습니다.<br>
DecisionTreeClassifier에 대한 내용은 나중에 정리하기로 하고, 일단 그냥 써보는겁니다!<br>

붓꽃 데이터 세트는 데이터가 합쳐져 있는 형태이기 때문에, 이를 분리하기 위해서 `sklearn.model_selection`에 있는 `train_test_split`을 사용하겠습니다.
