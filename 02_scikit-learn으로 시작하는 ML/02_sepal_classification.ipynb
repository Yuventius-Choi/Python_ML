# 붓꽃 품종 예측하기

첫 번째로 만들 모델은 붓꽃 데이터 세트로 품종을 분류하는 모델입니다.<br>
분류(Classification)는 지도학습(Supervised Learning) 방법 중의 하나로서,<br>
다양한 Feature 값과 Label 값으로 모델을 학습한 뒤, 테스트 데이터에서 Label을 예측하는 형태입니다.

Kaggle에서는 보통 데이터의 구성을 다음과 같이 제공합니다.
- train: 학습에 쓰일 데이터 셋 (Label 있음)
- val: 모델 검증용 데이터 셋 (Label 있음)
- test: 예측용 데이터 셋 (Label 없음)

따라서 다른 ML을 모델을 만들 때 해당 데이터 셋 구성으로 준비해두는 것이 좋습니다.<br>
물론 test 데이터 셋은 자율적으로 구성해도 됩니다.

## 데이터 로드 및 모델 선택

붓꽃 데이터는 Scikit-Learn 패키지에서 불러올 수 있습니다.<br>
데이터는 `sklearn.datasets` 안에 있습니다.

ML 알고리즘의 경우 DecisionTreeClassifier를 사용하겠습니다.<br>
DecisionTreeClassifier에 대한 내용은 나중에 정리하기로 하고, 일단 그냥 써보는겁니다!<br>

붓꽃 데이터 세트는 데이터가 합쳐져 있는 형태이기 때문에, 이를 분리하기 위해서 `sklearn.model_selection`에 있는 `train_test_split`을 사용하겠습니다.

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd

# 데이터 로드
iris = load_iris()

# IDE를 좋은 걸 쓰면 출력을 안하고도 Jupyter Variable을 볼 수 있습니다 :)
## 1. iris 데이터 셋은 data에 ndarray의 데이터가 존재합니다.
## 2. feature_names은 list형이고 ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'] 입니다.
## 3. target_names는 ndarray형이고 ['setosa' 'versicolor' 'virginica'] 입니다.
## 4. target은 target_names를 수치형으로 표시한 출력값입니다. [0, 1, 2]
iris_data = iris.data
iris_classes = iris.target_names
iris_label = iris.target
# print(f"iris target: {iris_label}")
# print(f"iris target_names: {iris.target_names}")

# 데이터를 DataFrame으로 변환
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
# label 값을 추가
iris_df["label"] = iris.target
iris_df.head(3)
```

train 데이터와 val 데이터를 나누기 위해 `train_test_split`을 사용합니다.<br>
`test_size=`를 지정하면 해당 데이터셋에서 어느 비율로 나눌지를 설정할 수 있습니다.<br>
`random_state=` 는 난수 Seed 설정으로, 어떤 환경에서든 해당 Seed로 구분하기 때문에 값이 항상 같아집니다.

```python
# 데이터 셋 설정
## X_train: 학습 데이터
## X_val: 검증 데이터
## y_train: 학습 Label
## y_val: 검증 Label
X_train, X_val, y_train, y_val = train_test_split(iris_data, iris_label, test_size=0.2, random_state=42)
```

ML 알고리즘 중 하나인 `DecisionTree`를 생성합니다.<br>
분류 문제에 해당하므로 Classifier를 사용합니다.

```python
# 결정트리 모델 생성
model_dt = DecisionTreeClassifier(random_state=42)
```

학습을 수행하기 위해서 `fit()` 함수를 사용합니다.

```python
model_dt.fit(X_train, y_train)
```

학습된 모델로 예측하려면 `predict()`를 사용합니다.

```python
y_pred = model_dt.predict(X_val)
```

학습된 모델을 평가하기 위해서는 여러가지 지표가 있지만 이건 나중에 정리하도록 하고 여기서는 Scikit-Learn의 `accuracy_score`를 사용합니다.<br>
정확도는 정답은 맞은 비율로서 수동으로 계산해도 되지만, 있는거 쓰는게 편하니까 불러와서 사용하도록 하겠습니다.<br>
정확도는 1.0에 가까울 수록 정확하다는 의미입니다.

```python
from sklearn.metrics import accuracy_score

# 정답을 앞에, 예측값을 뒤에 넣습니다.
print(f"accuracy: {accuracy_score(y_val, y_pred)}")
```

위의 과정과 같이 모든 모델의 학습 flow는 크게 이런식으로 진행됩니다.
1. 데이터 세트 분리
2. 모델 학습
3. 예측 수행
4. 평가

이 중간중간에 무엇을 더 명확하게 수행하는가가 모델의 성능을 결정합니다.
