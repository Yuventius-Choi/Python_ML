`describe()`가 수치형 데이터의 분포도를 표시해준다면, `value_counts()`는 범주형 데이터의 분포도를 확인하는데 사용합니다.<br>
해당 함수는 Series를 반환합니다.<br>

```python
pclass_counts = titanic_df["Pclass"].value_counts()
print(f"승객 등급 통계:\n{pclass_counts}")
```

#### 결측치 무시 - dropna()

`value_counts()`는 결측치를 포함한 결과값을 출력합니다.<br>
예시로 Embarked column을 확인해보겠습니다.

```python
print(f"타이타닉 데이터 갯수: {titanic_df.shape[0]}")

embarked_counts = titanic_df["Embarked"].value_counts()
print(f"embarked_counts shape: {embarked_counts.shape}")

tot_embarked_count = 0
for embarked_count in embarked_counts:
    tot_emabrked_count += embarked_count

print(f"embarked 데이터 갯수: {tot_emabrked_count}")
```

전체 데이터 891개에서 2개가 부족하다는 것을 알 수 있습니다.<br>
하지만 위의 `info()`를 보면 891개라고 표시되어 있어 데이터 갯수가 일치하지 않습니다.

이 2개의 데이터는 NaN 값이며, 다른 언어의 null과 같습니다.<br>
기본적으로 `value_counts()`는 NaN 값을 제외하고 표시하기 때문에 모든 값을 표시하고 싶다면 인자에 `dropna=False`를 사용해서 표시할 수 있습니다.

```python
embarked_with_nan_counts = titanic_df["Embarked"].value_counts(dropna=False)

print(embarked_counts)
print(embarked_with_nan_counts)
```

### DataFrame의 형변환

DataFrame은 `read_csv()`처럼 csv파일에서도 데이터를 읽어 초기화 할 수도 있지만<br>
Python의 list, dictionary, Numpy의 ndarray 등에서도 초기화 할 수 있습니다.<br>
물론 반대의 경우도 변환이 가능합니다.

API마다 입력인자로 받는 객체가 다르기 때문에 (특히 ndarray와 DataFrame 중 하나일 경우가 많습니다.)<br>
Pandas에서 이러한 형태의 형변환을 지원하게 됩니다.

#### ndarray, List, Dict -> DataFrame

위 collection과 DataFrame의 차이는 index가 명시적으로 선언되어 있는지의 차이가 있습니다.<br>
다만 DataFrame은 2차원 배열만을 저장할 수 있기 때문에, 다차원 ndarray는 변환할 수 없습니다.

1차원 배열을 DataFrame으로 변환하려면 column 지정이 필요하며 `columns=`를 인자로 넣어 지정합니다.

```python
# Data type을 명시적으로 표시하기 위한 내장 패키지
from typing import List, Dict
import numpy as np

col_name1: List = ["col1"]
list1: List = [1, 2, 3]

# list를 ndarray로 형변환
arr1: np.ndarray = np.array(list1)

# 리스트로 DataFrame 생성
# list1은 1차원이기 때문에 columns로 열 지정
list1_df: pd.DataFrame = pd.DataFrame(list1, columns=col_name1)
print(f"list1_df shape: {list1_df.shape}")
print(f"list_df:\n{list1_df}")

# ndarray로 DataFrame 생성
print(f"arr1 shape: {arr1.shape}")
# arr1도 1차원이기 때문에 columns로 열 지정
arr1_df: pd.DataFrame = pd.DataFrame(arr1, columns=col_name1)
print(f"arr1_df:\n{arr1_df}")
```

2차원 배열또한 column을 명시적으로 선언해줍니다.

```python
# 2x3 2차원 배열 생성
col_name2: List = ["col1", "col2", "col3"]
list2: List = [[1, 2, 3],
         [11, 12, 13]]

list2_df: pd.DataFrame = pd.DataFrame(list2, columns=col_name2)
print(f"list2_df:\n{list2_df}")

arr2: np.ndarray = np.array(list2)
print(f"arr2 shape: {arr2.shape}")

arr2_df: pd.DataFrame = pd.DataFrame(arr2, columns=col_name2)
print(f"arr2_df:\n{arr2_df}")
```

Dictionary의 경우 Key를 컬럼명으로, Value는 데이터로 변환됩니다.<br>
ndarray는 Dictionary 변환을 지원하지 않습니다.

```python
dict1: Dict = {"col1": [1, 2, 3, 4],
         "col2": [11, 12, 13, 14],
         "col3": [21, 22, 23, 24]}

dict1_df = pd.DataFrame(dict1)
print(f"dict1_df:\n{dict1_df}")
```

#### DataFrame -> ndarray, List, Dict

반대의 경우도 당연히 가능합니다.<br>
특히 DataFrame을 ndarray로 변환하는 과정은 실제로 가장 많이 쓰이게 될 기능이라 익숙해지면 좋습니다.<br>
DataFrame의 `values`로 컬럼과 index를 제외한 값을 shape에 맞게 가져올 수 있습니다.

List의 경우 ndarray의 `tolist()` 함수로 변환하여 총 2번 바꾸게 됩니다.

Dictionary의 경우 DataFrame에서 `to_dict()` 함수를 사용하고, 인자로 'list'를 넣어 해당 값을 list형으로 가져옵니다.

```python
# dict1_df는 4x3 형태의 2차원 배열
## ndarray 변환
arr3: np.ndarray = dict1_df.values
print(f"arr3 type: {type(arr3)}")
print(f"arr3 shape: {arr3.shape}")
print(f"arr3:\n{arr3}")

## List 변환
list3: List = dict1_df.values.tolist()
print(f"list3 type: {type(list3)}")
print(f"list3:\n{list3}")

## Dictionary 변환
dict3: Dict = dict1_df.to_dict("list")
print(f"dict3 type: {type(dict3)}")
print(f"dict3:\n{dict3}")
```

## DataFrame의 Column 추가 및 수정

DataFrame은 mutable한 객체이므로, 컬럼이나 데이터 수정이 자유롭습니다.

### Column 추가

`[]`로 Column을 추가할 수 있으며, 초기값을 지정해 줄 수 있습니다.

```python
# 원본 유지를 위해 복사합니다.
## deep은 깊은 복사를 수행합니다.
titanic_cp_df = titanic_df.copy(deep=True)

# Age_0 컬럼을 초기값 0으로 생성
titanic_cp_df["Age_0"] = 0
# 기존 컬럼을 사용하여 생성
titanic_cp_df["Age_by_10"] = titanic_cp_df["Age"] * 10
titanic_cp_df["Family_No"] = titanic_cp_df["SibSp"] + titanic_cp_df["Parch"] + 1

titanic_cp_df.head(3)
```

#### 데이터 일괄 수정

해당 컬럼에 접근하여 여러 연산으로 컬럼 값을 덮어씌울 수 있습니다.

```python
titanic_cp_df["Age_by_10"] = titanic_cp_df["Age_by_10"] + 100

titanic_cp_df.head(3)
```

### DataFrame 데이터 삭제

`drop()` 함수를 사용하여 행이나 열을 삭제합니다.<br>
주로 쓰이는 인자는 다음과 같습니다.
- `axis=` :<br>
대부분의 경우 행을 지우는 경우가 없기 때문에 (데이터 n건을 삭제하는 효과) 필요없는 column을 삭제하기 위해서 `axis=1`를 사용하게 됩니다.<br>
이상치에 해당하는 데이터를 삭제하는 경우 `axis=0`을 사용합니다.
- `inplace=` :<br>
데이터 삭제는 risk가 있는 작업일 수 밖에 없기에 기본적으로는 삭제된 DataFrame을 복사해서 반환합니다.<br>
`inplace=True`가 설정되면, 원본에서 직접 삭제하게 됩니다.
- `labels=` :<br>
여러 행, 열을 삭제하고 싶을 때 리스트로 index나 column name을 지정하여 사용합니다.

위에서 생성한 모든 컬럼을 삭제하도록 하겠습니다.

```python
# column 삭제
need_drop_cols: List = ["Age_0", "Age_by_10", "Family_No"]
# inplace가 True 이므로, 반환값은 None이 됩니다.
drop_result: pd.DataFrame = titanic_cp_df.drop(need_drop_cols, axis=1, inplace=True)
print(f"drop_result:\n{drop_result}")
print("원본:")
titanic_cp_df.head(3)
```

```python
# row 삭제
need_drop_rows = [0, 1, 2]
titanic_cp_df.drop(labels=need_drop_rows, axis=0, inplace=True)
titanic_cp_df.head(3)
```

## Index 객체

Index는 DataFrame에서 PK의 역할을 하는 객체입니다.<br>
`index` 로 따로 배열로 추출할 수 있습니다.<br>
`values`로 ndarray 형태로 추출할 수 있어 numpy의 API를 모두 사용할 수 있습니다.

index 자체는 Immutable 객체이기 때문에 직접 변경이 불가능합니다.

```python
# cp 원복
titanic_cp_df: pd.DataFrame = titanic_df.copy(deep=True)
# index 배열 추출
indexes: pd.RangeIndex = titanic_cp_df.index
print(f"indexes type: {type(indexes)}")
print(f"indexes shape: {indexes.shape}")
print(f"indexes:\n{indexes}")

# array로 변환
index_arr: np.ndarray = indexes.values
print(f"index_arr type: {type(index_arr)}")
print(f"index_arr shape: {index_arr.shape}")
print(f"index arr:\n{indexes.values}")
```

`RangeIndex` 객체는 ndarray처럼 slicing이 가능합니다.

```python
# RangedIndex slicing
print(indexes[:5].values)
# ndarray slicing
print(indexes.values[:5])
```

Series 객체는 Index 객체를 포함하지만, 연산에는 사용되지 않습니다.<br>
따라서 Index는 식별용(참조용)으로만 사용하게 됩니다.

만약 index값이 연속된 숫자형태가 아니라면, 데이터 sort기준으로 새로 인덱스를 부여해줄 수 있습니다.<br>
`reset_index()`로 새로 생성할 수 있고 `inplace=True`가 기본값입니다.<br>
`inplace=False`인 경우, 기존 index를 컬럼으로 추가하고 새로운 연속 숫자형 index를 생성합니다.

```python
titanic_cp_df = titanic_cp_df.reset_index(inplace=False)
titanic_cp_df.head(3)
```

이전 Pclass Series를 가져왔을 때 [3, 1, 2]로 구분되어 있었는데 여기서 `reset_index()`를 실행하면 기존 [3, 1, 2]가 column으로 바뀌고,<br>
연속 [0, 1, 2] 값이 index로 지정되게 됩니다.<br>
`inplace=False` 라면 column이 추가되었으므로 Data type은 DataFrame으로 자동으로 바뀝니다.

```python
pclass_series = titanic_cp_df["Pclass"].value_counts()
print(f"pclass_series type: {type(pclass_series)}")
print(f"pclass_series:\n{pclass_series}")

# index 새로 지정
pclass_df = pclass_series.reset_index(inplace=False)
print(f"pclass_df type: {type(pclass_df)}")
print(f"pclass_df:\n{pclass_df}")
```

## 데이터 셀렉션 및 필터링

Pandas는 Numpy와 상당히 유사한 셀렉션 및 필터링 기능을 가지고 있지만, 다른 부분도 분명히 있으므로 혼동되기 쉽습니다.<br>
Numpy의 데이터 핸들링은 데이터 분석용으로는 한계가 있기 때문에, Pandas에서는 Numpy의 기능을 계승하거나 혹은 완전히 다른 기능을 제공하기도 합니다.<br>
DataFrame과 Series끼리도 달라지는 부분이 있어서 주의가 필요합니다.

Numpy에서는 `[]` 연산자를 사용하여 다음과 같은 기능을 수행했습니다.
- 단일 값 추출
- Slicing
- Fancy Indexing
- Boolean Indexing

하지만 Pandas에서는 이러한 작업을 `loc[]`, `iloc[]` 연산자를 통해 수행하게 됩니다.<br>

먼저 Numpy의 `[]`와 Pandas의 `[]`의 차이를 확인해보도록 하겠습니다.

### DataFrame의 `[]`연산자

Numpy에서 `[]`연산자는 행의 위치, 열의 위치, Slicing 범위 등을 지정해서 가져올 수 있었습니다.<br>
하지만 DataFrame의 `[]`연산자는 column명, 또는 index로 변환 가능한 표현식(statement)입니다.<br>

```python
# df 원복
titanic_cp_df = titanic_df.copy(deep=True)

print(f"단일 column 데이터 추출:\n{titanic_cp_df['Pclass'].head(3)}")
print(f"여러 column 데이터 추출:\n{titanic_cp_df[['Survived', 'Pclass']].head(3)}")

# index로 접근시 에러 발생
print(f"index 로 접근:\n{titanic_cp_df[0]}")
```

인덱스 형태로 변환 가능한 표현식은 다음 요소만 가능합니다.
- Slicing
- Boolean Indexing

거의 대부분이 Boolean Indexing을 활용하므로, 익숙해지면 좋습니다.

```python
# Slicing
# head(3)와 같습니다.
titanic_cp_df[0:3]
```

```python
# Boolean Indexing
# Pclass가 3인 최상위 3개
titanic_cp_df[titanic_cp_df["Pclass"] == 3].head(3)
```

### `iloc[]` 연산자

`iloc[]`의 경우 인덱스 기반으로 DataFrame의 값을 참조하는 함수입니다.<br>
따라서 숫자가 아닌 값으로 참조하면 에러가 발생합니다.

### Aggregation

### groupby()

RDBMS의 `group by`와 비슷하지만 다른 면이 있기 때문에 주의해야 합니다.<br>
DataFrame에서 `groupby()`를 호출하게 되면 DataFrameGroupBy라는 다른 형태의 DataFrame을 반환하게 됩니다.

```python
titanic_groupby = titanic_cp_df.groupby(by="Pclass")
print(type(titanic_groupby))
```

DataFrame에서 `groupby()`를 수행하면 지정한 column을 제외하고 aggregation이 적용됩니다.<br>
RDBMS에서는 모든 column에 aggregation을 적용해야 하지만 `SELECT count(PassengerId), count(Survived),...`<br>
DataFrame은 일괄 적용 시킨다는 점이 다르다고 할 수 있습니다.

```python
# Aggregation 일괄 적용
titanic_groupby = titanic_cp_df.groupby(by="Pclass").count()
titanic_groupby
```

```python
# 특정 column에만 Aggregation 적용
titanic_groupby = titanic_cp_df.groupby(by="Pclass")[["PassengerId", "Survived"]].count()
titanic_groupby
```

만약 하나의 컬럼에 여러 aggregation을 적용하고 싶거나 여러 컬럼에 각각의 aggregation을 적용하고 싶다면 `agg()` 함수를 사용합니다.<br>
`agg()` 함수는 aggregation list와 "column": "aggregation" 형태의 dictionary를 받을 수 있습니다.<br>
확실히 RDBMS보다는 복잡한 과정을 가지고 있습니다.

```python
# aggregation list
## Pclass 별 Age의 최소, 최대 출력
print("Pclass별 Age 최대 최소")
titanic_cp_df.groupby("Pclass")["Age"].agg([max, min])
```

```python
# aggregation dictionary
agg_format = {"Age": "max", "SibSp": "sum", "Fare": "mean"}
print("Pclass 별 Age 최대, SibSp 총합, Fare 평균")
titanic_cp_df.groupby("Pclass").agg(agg_format)
```

## 결측치 처리

결측치는 cell에 데이터가 없는 경우를 의미하며, Numpy의 NaN으로 표시합니다.<br>
ML은 이 결측치를 자동으로 처리하지 않으므로 다른 값으로 채워넣어야 합니다.<br>
또한 이 결측치는 Aggregation 시 제외가 되므로 데이터 신뢰도가 낮아지는 현상이 있을 수 있습니다.

### 결측치 여부 확인 - isna()

isna()는 해당 데이터가 NaN인지를 알려주는 함수입니다.<br>
DataFrame에 적용하면 모든 column의 값이 NaN인지 아닌지를 [True/False]로 반환합니다.

```python
titanic_cp_df.isna().head()
```

```python
# 결측치 갯수 확인
titanic_cp_df.isna().sum()
```

### 결측치 대체 - fillna()

결측치는 일일히 넣어줄 필요없이 `fillna()` 함수로 일괄 처리할 수 있습니다.<br>
이 함수 역시 `inplace` 인자가 존재하며, 기본값은 False 입니다.

```python
# Cabin column의 결측치를 모두 C000으로 대체합니다.
titanic_cp_df["Cabin"] = titanic_cp_df["Cabin"].fillna("C000")
titanic_cp_df.head(3)
```

```python
# 1. Age의 결측치를 모두 평균값으로 대체
# 2. Embarked의 결측치를 모두 S로 대체
age_mean = titanic_cp_df["Age"].mean()
print(f"Age의 평균값: {age_mean}")

titanic_cp_df["Age"].fillna(age_mean, inplace=True)
titanic_cp_df["Embarked"].fillna("S", inplace=True)

# 모든 결측치가 처리되었습니다.
titanic_cp_df.isna().sum()
```

## lambda 식으로 데이터 가공

lambda 식은 대부분의 경우 익숙하지 않은 표현식이지만, 언어와 상관없이 사용되는 개념입니다.<br>
Python에서의 lambda 문법이긴 하지만 DataFrame에서는 `apply()`로 lambda식을 활용할 수 있습니다.

lambda에 대해 모르겠으면 Python을 다시 공부하고 오시면 됩니다. ^^

```python
# 예시 함수 선언
def get_square(a):
    return a ** 2

print(f"3의 제곱은 {get_square(3)}")
```

```python
lambda_square = lambda x: x ** 2
print(f"3의 제곱은 {lambda_square(3)}")
```

Python에서는 여러 값에 일괄적용하기 위해서는 `map()`을 사용해야 합니다.

```python
list1 = [1, 2, 3]
list1_squares = list(map(lambda_square, list1))
print(list1_squares)
```

Pandas에서는 이러한 lambda 식을 적용하기 위해 `apply()` 함수를 사용합니다.

```python
titanic_cp_df["Name_len"] = titanic_cp_df["Name"].apply(lambda x: len(x))
titanic_cp_df[["Name", "Name_len"]].head(3)
```

Python의 lambda식 형식에 맞다면, 복잡한 구문도 적용이 가능합니다.<br>
다만 너무 길어지면 가독성이 매우 떨어지므로, 별도의 함수로 구현하는게 좋습니다.

```python
# 18세 이하라면 Child, 그 이외는 Adult
titanic_cp_df["Child_Adult"] = titanic_cp_df["Age"].apply(lambda x: "Child" if x <= 18 else "Adult")
titanic_cp_df[["Age", "Child_Adult"]].head(8)
```

```python
titanic_cp_df["Child_Adult"].value_counts()
```

```python
# 함수로 lambda 적용
def get_category(age):
    cat = ""
    if age <= 5: cat = "Baby"
    elif age <= 12: cat = "Child"
    elif age <= 18: cat = "Teenager"
    elif age <= 25: cat = "Student"
    elif age <= 35: cat = "Young Adult"
    elif age <= 60: cat = "Adult"
    else: cat = "Elderly"

    return cat

titanic_cp_df["Age_cat"] = titanic_cp_df["Age"].apply(lambda x: get_category(x))
titanic_cp_df["Age_cat"].value_counts()
```
